# LET DO SOME VISUALIZATION with Matplotlib

```python
x = np.linspace(0,10,100)
y = np.linspace(0,50,100)
```

```python
%matplotlib inline
plt.plot(x,y,linewidth=5,label='Some raw data') 
plt.show()
```
