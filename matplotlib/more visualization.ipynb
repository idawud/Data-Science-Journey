```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

# Functional Way

```python
plt.plot(x, y, color='indigo', linewidth=2.5)
plt.xlabel(' x label')
plt.ylabel(' y label')
plt.title(' Title')
```

```python
# 'r' == red & 'b' == blue
# subplot() argument number of rows and columns and figure number
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# ObjectOriented Way

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.525,0.5,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x, 'gx', linestyle='dotted', linewidth=2, markersize=10)
axes2.set_title('Smaller Plot')
```
